Análisis y corrección de tipos.

pasar event_time a datetime
Pasamos event_time a datetime.

TRUCO PRO: pd.to_datetime() puede tardar mucho en ejecutarse en datasets grandes.

Pero por algún motivo si dividimos la cadena de la fecha en sus partes y la volvemos a juntar y después transformamos a datetime especificándole el formato exacto funciona MUCHO más rápido.

Te voy a enseñar las dos formas:

la fácil y tradicional con pd.to_datetime: puedes usar esta si no quieres complicarte con la función
la avanzada creando una función para hacer lo que he comentado
Te recuerdo el link para los códigos de los formatos:

https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior

En mi equipo con 32GB de RAM la forma tradicional tardó más de 5 minutos, mientras que la avanzanda menos de 10 segundos.

```python
#forma tradicional, la dejo comentada porque voy a utilizar la otra
#df.event_time = pd.to_datetime(df.event_time)
```

```python
#forma avanzada crando una función
#a esta función hay que pasarle la variable fecha y el formato en el que está

def datetime_rapido(dt,formato):

    def divide_fecha(fecha):
        division = fecha.split()
        date = division[0]
        time = division[1]
        cadena = date + ' ' + time
        return cadena

    resultado = pd.to_datetime(dt.apply(lambda x: divide_fecha(x)), format = formato)

    return resultado
```

```python
#Ejecutamos la función

formato = '%Y-%m-%d %H:%M:%S'

df.event_time = datetime_rapido(df.event_time, formato)
df.info()
```

Para solucionarlo tenemos que TIPIFICAR las Variables y ponerlas asi en la misma escala

TIPIFICAR: Tomar cada dato de las distribución, restarle la media y dividir el resultado por la desviación típica

```python
#Creamos FUNCIÓN TIPIFICADA:

def tipificar(variable):
    
    media = variable.mean()
    dt = variable.std()
    return(variable.apply(lambda x: (x - media) / dt))
```

### Cual es la supervivencia de los Clientes?

Hacemos un ANÁLISIS DE SUPERVIVENCIA a partir de una ANÁLISIS DE COHORTES para hacer una medición a lo largo del tiempo de los clientes que captamos, para saber cuantos nos siguen comprando meses despues.

Con el ANÁLISIS DE COHORTES replicamos varias veces el ANÁLISIS DE SUPERVIVENCIA para comprobar que es un análisis estable. Cuantos mas meses de histórico tengamos mas COCHORTES podemos hacer y mas fiables serán los análisis

```python
#Preparamos los datos en el formato que necesitamos
#Necesitamos a cada cliente en las filas y en las columnas los meses en los que cada cliente estuvo activo y realizó compra

c = df.loc[df.evento == 'purchase',['usuario','mes']]
c
```

```python
#Pasamos MESES  a COLUMNAS haciendo un CONTEO del número de compras que tuvo ese cliente en cada mes

c = pd.crosstab(c.usuario,c.mes)
c
```

```python
#Sacamos el usuario del INDEX

c = c.reset_index()
c
```

```python
#Renombramos las columnas en COHORTES 
#Eliminamos la columna USUARIO, ya que sabemos que cada usuario es una FILA

c.columns = ['usuario','c4','c5','c1','c2','c3']
c.drop(columns = 'usuario', inplace= True)
c
```

Ya tenemos la estructura que necesitamos en nuestro df

**PROCESO PARA CALCULAR COHORTES:**

**1. Seleccionar para cada COHORTE unicamente los datos que aplican a esa cohorte; Seleccionar la base**

**2. Binarizar, escala 1/0**

**3. Hacer MEDIA d elos clientes que compran en cada mes(supervivientes)**

**COHORTE2**

```python
# 1.Clientes que NO estaban en el mes1 y SI compran en el mes2(base)

c2 = c.loc[(c.c1 == 0) & (c.c2 > 0)]
c2
```

```python
# 2. Binarizar

def binarizar(variable):
    variable = variable.transform(lambda x: 1 if (x > 0) else 0)
    return(variable)

c2_b = c2.apply(binarizar)
c2_b
```

```python
# 3. Calcular la media de cada mes
#Devuelve la proporción al estar en binario

c2_b = c2_b.mean().sort_index()
c2_b
```

Devuelve la PROPORCIÓN de clientes que sigue vivo los meses posteriores para la COHORTE2

- Solo el 10% de todos los clientes nuevos del MES2 sigue comprando el MES3
- Solo el 10% siguen comprando el MES4
- Solo el 8% sigue comprando el MES5

**COHORTE3 Y 4**

```python
#COHORTE3
# 1. Seleccionar la base
c3 = c.loc[(c.c2 == 0) & (c.c3 > 0)]

# 2. Binarizar

c3_b = c3.apply(binarizar)

# 3. Media

c3_b = c3_b.mean().sort_index()
c3_b ['c1'] = 0
c3_b
```

```python
#COHORTE4
# 1. Seleccionar la base
c4 = c.loc[(c.c3 == 0) & (c.c4 > 0)]

# 2. Binarizar

c4_b = c4.apply(binarizar)

# 3. Media

c4_b = c4_b.mean().sort_index()
c4_b ['c1'] = 0
c4_b ['c2'] = 0
c4_b
```

```python
#Unimos los datos de las 3 COHORTES en un mismo DataFrame

cohortes = pd.DataFrame({'c2': c2_b,
              'c3': c3_b,
              'c4': c4_b})
cohortes
```

```python
#Eliminamos c1 que no aporta ningun dato (todo 0.00)

cohortes = cohortes.drop(index = 'c1').T
cohortes
```

### Construyendo un Sistema de Recomendación

#### Crear DataFrame con KPIs de interés:

En este caso usaremos lo que se llama un kpi implícito, que será el número de veces que los productos han sido comprados por el mismo usuario.

Kpis explícitos sería por ejemplo estrellas o puntuaciones del 1 al 10.

Dado que este es una algoritmo que tarda en calcularse vamos a reducir el problema y calcularlo solo para los 100 productos más vendidos.

Además, por motivos didácticos vamos a hacer el paso a paso manualmente.

Primero calculamos un maestro con los 100 productos más vendidos.

```python
mas_vendidos = prod.sort_values('purchase', ascending = False).producto[0:100]
mas_vendidos
```

```python
#Creamos el df temporal
temp = df.loc[df.producto.isin(mas_vendidos)]
temp
```

```python
#Filtramos por 'purchase', agrupamos por 'usuario' y 'producto', pasamos 'producto' a columnas 

usuario_item = temp.loc[temp.evento == 'purchase'].groupby(['usuario','producto']).size().unstack(level = 1).fillna(0)
usuario_item
```

#### Reducir la dimensión (opcional):

Nos devuelve una MATRIZ DISPERSA O SPARSE

Sería conveniente reducir la dimensión con técnica SVD

#### Seleccionar la Métrica de Distancia:

En este caso vamos a seleccionar la MÉTRICA EUCLIDEA

La operativizamos mediante la función spatial.distance.euclidean de Scipy

```python
from scipy import spatial
```

#### alcular la Matriz Item-Item:

```python
#Creamos una función personalizada

def recomendador(dataframe):
    
    def distancia(producto):
        
        return(dataframe.apply(lambda x: spatial.distance.euclidean(x, producto)))
    
    return(dataframe.apply(lambda x: distancia(x)))
```

```python
item_item = recomendador(usuario_item)
item_item
```

#### Crear la lógica de Priorización:

Ya tenemos listo el recomendador.

Lo que tendríamos que hacer es una llamada a esta tabla cada vez que un usuario mire un producto o lo meta en el carrito.

Pero para que sea más efectivo podríamos usar toda la info acumulada de la sesión o incluso de todo el usuario si está logado.

Eso significa que necesitamos un sistema para recomendar productos tanto si el input es de un solo producto como de varios.

Y que a la vez devuelva varias recomendaciones, para cubrir todos los "huecos" de recomendación que nuestra web pudiera tener.

Aplicaremos un algoritmo muy sencillo que hará:

1. Crear un array con los productos de entrada para extraer sus vectores de la matriz item-item

2. Calcular la suma de distancias de todos los productos

3. Quitarse a ellos mismos para no autorecomendarse.
 
4. Devolver los 10 con menor distancia

```python
#En el caso de varios productos vendrá del servidor web como una cadena separada con punto y coma
def priorizador(productos,devuelve = 10):
    #crear array con productos de entrada
    array = np.int64(productos.split(';'))
    
    #extraer sus vectores de la matriz total
    matriz = item_item[array]
    
    #calcular la suma de distancias
    suma_distancias = matriz.agg(sum,axis = 1)
    
    #eliminar los productos input
    suma_distancias = suma_distancias.loc[~suma_distancias.index.isin(list(array))]
    
    #Devolver los 10 con menor distancia
    return(suma_distancias.sort_values()[0:devuelve])
```

```python
#Comprobamos cómo funciona si le pasamos un producto

priorizador('4497')
```

```python
#Comprobamos cómo funciona si le pasamos varios productos

priorizador('4497;4600;4768')
```
